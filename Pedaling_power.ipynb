# Title : Pedaling power processing 

## Application to Phyling' s data 

```python
import numpy as np
import pandas as pd
import tkinter as tk 
import math as mt
import Pedaling_torque as pedaling_torque
import Pedaling_cadence as pedaling_cadence
```

## I. Read file

```python
def SelectFile() :
    root = tk.Tk() 
    root.withdraw()
    file_path = tk.filedialog.askopenfilename(
        title = "Select a file",
        filetypes = ("CSV Files", ".* csv")
    )
    global data_Phyling_table
    data_Phyling_table= pd.read_csv(file_path)
```

```python
data_Phyling_table[1] = []
```

# II. Pedaling power process

```python
if pd.columns.str.contains("Maxi269.Couple_d").any() and pd.columns.str.contains("Maxi269.Cadence").any() in data_Phyling_table : 
        if len(data_Phyling_table["Maxi269.Cadence"]) == len(data_Phyling_table["Maxi269.couple_d"]) :
                for i in range(len(data_Phyling_table["Maxi269.Cadence"])) and range(len(data_Phyling_table["Maxi269.couple_d"])) :
                        Right_power = pedaling_torque.Right_torque[i] * pedaling_cadence.rad_per_second[i]
                        data_Phyling_table.append(Right_power)
                       
```

```python
if pd.columns.str.contains("Maxi269.Couple_g").any() and pd.columns.str.contains("Maxi269.Cadence").any() in data_Phyling_table : 
        if len(data_Phyling_table["Maxi269.Cadence"]) == len(data_Phyling_table["Maxi269.couple_g"]) :
                for i in range(len(data_Phyling_table["Maxi269.Cadence"])) and range(len(data_Phyling_table["Maxi269.couple_g"])) :
                        Left_power = pedaling_torque.Left_torque[i] * pedaling_cadence.rad_per_second[i]
                        data_Phyling_table.append(Left_power)   
```

## III. Output results file 

```python
data_Phyling_table.to_csv("data_Phyling_table.csv", index=False)
```
