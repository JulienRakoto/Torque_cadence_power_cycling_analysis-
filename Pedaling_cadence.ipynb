# Title : Pedaling cadence processing 

## Application to Phyling' s data 

```python
import numpy as np
import pandas as pd
import tkinter as tk 
```

## I. Read file

```python
def SelectFile() :
    root = tk.Tk() 
    root.withdraw()
    file_path = tk.filedialog.askopenfilename(
        title = "Select a file",
        filetypes = ("CSV Files", ".* csv")
    )
    global data_Phyling_table
    data_Phyling_table= pd.read_csv(file_path)
```

```python
data_Phyling_table[1] = []
```

# II. Pedaling cadence process

```python
if pd.columns.str.contains("Maxi269.Theta").any() and pd.columns.str.contains("T").any() in data_Phyling_table :
        if len(data_Phyling_table["Maxi269.Theta"]) == len(data_Phyling_table["T"]) :
                for i in range(len(data_Phyling_table["Maxi269.Theta"])) and range(len(data_Phyling_table["T"])) :
                        degree_per_second = np.diff(data_Phyling_table["Maxi269.Theta"][i]) / np.diff(data_Phyling_table["T"][i]) # expressed as degree per second
                        cycle_per_minute = degree_per_second / 360  # expressed as cycle per minute
                        rad_per_second = cycle_per_minute * ((2 * np.pi) / 60) # expressed as radian per second
                        data_Phyling_table.append(rad_per_second)





                       
```

## III. Output results file 

```python
data_Phyling_table.to_csv("data_Phyling_table.csv", index=False)
```
